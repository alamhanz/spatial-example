# Read Shapefile

```python
import hvplot.pandas
import pandas as pd
import geopandas as gpd
import random
import numpy as np
```

```python
path_shp = '../../batas-administrasi-indonesia/Kel_Desa/Kel_Desa/'
```

## Read

```python
fp = path_shp+'Kel_Desa.shp'
map_df = gpd.read_file(fp)
```

```python
map_df.plot(figsize=(20,10))
```

```python
map_df.sample(10)
```

```python
map_df[map_df['NAME'].str.lower().str.contains('blitar')]
```

```python
map_blitar = map_df[(map_df['KAB_KOTA']=='Blitar')|(map_df['KAB_KOTA']=='Kota Blitar')]
map_jatim = map_df[map_df['PROVINSI']=='Jawa Timur']
```

```python
## random number
# map_blitar['rand_val'] = [random.randint(2,25) for i in range(map_blitar.shape[0])]
map_blitar['rand_val'] = np.where(map_blitar['KAB_KOTA']=='Kota Blitar',1,0)
```

```python
map_blitar.plot(figsize=(20,10))
```

```python
map_blitar.hvplot(geo=True, tiles="OSM", c='rand_val',width=1200, height=800)
```

```python
map_blitar[map_blitar['KEL_DESA']=='Bendosari'].hvplot(geo=True, tiles="OSM", c='rand_val',width=1200, height=800)
```
